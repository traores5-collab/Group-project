*sigmoid function*

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

*Cost function*

```python
def costFunction(theta, X, y):
    m = np.size(y)
    hprime= X.dot(theta)
    h = sigmoid(hprime)
    j=-1/m*(np.log(h).T.dot(y)+np.log(1-h).T.dot(1-y))
    j_2=(-1/m)* np.sum(y*np.log(h)+(1-y)*np.log(1-h))
     
               
    if np.isnan(j[0]):
        return(np.inf)
    return(j[0])
```

*Gradient descent*

```python

def gradient(theta, X, y):
    m = np.size(y)
    hprime=X.dot (theta)
    h = sigmoid (hprime)
    
    grad = (-1/m)*X.T.dot(h-y)
    #grad2=(-1/m)*

    return(grad.flatten())
```

```python
initial_theta = np.zeros((X.shape[1],1))
cost = costFunction(initial_theta, X, y)
grad = gradient(initial_theta, X, y)
print('Cost: \n', cost)
print('Grad: \n', grad)
```

```python
def gradientDescent(X, y, theta=None, alpha=0.01, num_iters=50):
    if theta is None:
        theta = np.zeros((X.shape[1], 1))  # Automatically match number of features

    m = len(y)
    j_history = np.zeros(num_iters)

    for i in range(num_iters):
        h = sigmoid(X.dot(theta))
        gradient = (1 / m) * X.T.dot(h - y)
        theta = theta - alpha * gradient
        j_history[i] = (-1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))

    return theta, j_history
```

```python
theta, j_history = gradientDescent(X, y)
```

*model prediction*

```python
def predict_class(user_input, theta):
    if user_input.ndim == 1:
        user_input = user_input.reshape(1, -1)

    # Pas de biais ici
    prediction = sigmoid(user_input.dot(theta))
    return int((prediction >= 0.5)[0][0])
```

```python

mon_input = np.array([45, 190])  
résultat = predict_class(mon_input, theta)
print("Prédiction :", résultat)

          
```
